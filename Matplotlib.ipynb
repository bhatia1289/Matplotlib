```python
#line plot
#whenever you want to depict the realtion beteen to variables in the form of trend
#mostly in x axis the date -time column is there
x = [1,2,3,4,5] #asume x is no of days
y = [5,2,7,8,1] #asume stock price
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, color= 'red')
plt.title('line chart to understand the syntax')
plt.xlabel('day interval as x points')
plt.ylabel('price as y points')
plt.show() #no metadata
```

```python
plt.plot(x,y, color= 'red', marker = 'o')
plt.title('line chart to understand the syntax')
plt.xlabel('day interval as x points')
plt.ylabel('price as y points')
plt.show() #no metadata
```

```python
plt.plot(x,y, color= 'red',marker = 'o', linestyle = '--')
plt.title('line chart to understand the syntax')
plt.xlabel('day interval as x points')
plt.ylabel('price as y points')
plt.show() #no metadata
```

```python
plt.plot(x,y, color= 'red',marker = 'o', linestyle = '--',  linewidth = 4)
plt.title('line chart to understand the syntax')
plt.xlabel('day interval as x points')
plt.ylabel('price as y points')
plt.show() #no metadata
```

```python
plt.plot(x,y, color= 'red',marker = 'o', linestyle = '--',  linewidth = 4, markersize =15)
plt.title('line chart to understand the syntax')
plt.xlabel('day interval as x points')
plt.ylabel('price as y points')
plt.show() #no metadata
```

```python
#adding grid
plt.plot(x,y, color= 'red',marker = 'o', linestyle = '--',  linewidth = 4, markersize =15)
plt.title('line chart to understand the syntax')
plt.xlabel('day interval as x points')
plt.ylabel('price as y points')
plt.grid()
plt.show() #no metadata
```

```python
x= [1,2,3,4,5]
y1 =[5,7,1,8,9]
y2= [2,6,2,2,5]

plt.plot(x, y1,color='red', marker ='o', linestyle = '--', linewidth =3, markersize =6,label = "y1 as tata motors")
plt.plot(x, y2,color='blue', marker ='o', linestyle = '--', linewidth =3, markersize =6,label = "y2 as tesla")

plt.title('multiple lines in one plot')

plt.xlabel('Days')
plt.ylabel('stock price')
plt.legend()
plt.show()
```

```python
x=  np.linspace(0,10,100)
```

```python
plt.plot(x, np.sin(x), '--')
```

```python
plt.plot(x, np.sin(x), '.')
```

```python
#how to make 2 plots in one go
plt.figure() #initialising a page to make plot
plt.subplot(1,2,1) #(row, column , part)
plt.plot(x, np.sin(x))

plt.subplot(1,2,2)
plt.plot(x, np.cos(x))

plt.show()
```

```python
#how to make 2 plots in one go
plt.figure() #initialising a page to make plot
plt.subplot(2,1,1) #vertical
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))

plt.show()
```

```python
# make 4 subplots
plt.figure()
plt.subplot(2,2,1)
plt.plot(x, np.sin(x))
plt.subplot(2,2,2)
plt.plot(x, np.cos(x))
plt.subplot(2,2,3)
plt.plot(x, np.tan(x))
plt.subplot(2,2,4)
plt.plot(x, np.sin(x+1))

plt.show()
```

```python
#stacked histogram ( two histograms together)

data1 = np.random.normal(0,1,5000)
data2 = np.random.normal(3,2,5000)

plt.hist([data1,data2], bins=20, color= ['r','b'], label=['dist1', 'dist2'])
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('stacked histogram')
plt.show()
```

```python
#scatterplot >> to understand the relationship between two features
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y, c= 'r', marker='X')
```

```python
#bar plt >> to see distribution of categorical variables
x = ['a', 'b', 'c', 'd', 'e'] #name of students
y = [4,6,8,9,2] # no of days presnt in the class

plt.bar(x,y, color ='r') #vertical plot
```

```python
#bar plt >> to see distribution of categorical variables
x = ['a', 'b', 'c', 'd', 'e'] #name of students
y = [4,6,8,9,2] # no of days presnt in the class

plt.barh(x,y, color ='r') #horizontal plot
```

```python
#pie chart >> to understand the percent distribution of data

langs =['C', 'Python', 'Java']
students = [23,17,19]

plt.pie(students, labels = langs, colors = ['orange', 'blue', 'g'], autopct='%1.1f%%')
plt.title("piechart")
plt.show()
```

```python
#pie chart >> to understand the percent distribution of data
explode = (0.0, 0.1,0.1)
langs =['C', 'Python', 'Java']
students = [23,17,19]

plt.pie(students, labels = langs, colors = ['orange', 'blue', 'g'], autopct='%1.1f%%', explode = explode)
plt.title("piechart")
plt.show()
```

```python
#pie chart >> to understand the percent distribution of data
explode = (0.0, 0.1,0.1)
langs =['C', 'Python', 'Java']
students = [23,17,19]

plt.pie(students, labels = langs, colors = ['orange', 'blue', 'g'], autopct='%1.1f%%', explode = explode, shadow = True)
plt.title("piechart")
plt.show()
```

```python
#area plot >> variables changed over time
years = [2021 , 2018 , 2022, 2023, 2017]
category1 = [20, 25, 30, 35, 40]
category2 = [10, 15, 20, 25, 30]
category3 = [5, 10, 15, 20, 25]

plt.stackplot(years, category1, category2, category3, labels = ['cat1', 'cat2', 'cat3'], colors = ["skyblue", "lightgreen","orange" ])
plt.xlabel('years')
plt.ylabel('values')
plt.title('area plot')
plt.legend()
plt.show()
```
